# Import library

```python
import pandas as pd
from matplotlib import pyplot as plt
from statsmodels.tsa.seasonal import STL
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

import cslib.cslib as cslib

%matplotlib inline
```

# Read Data

```python
revenue_all = pd.read_csv('../data/ingest/revenue-all.csv')
```

# Plot revenue over time

```python
revenue_all.plot(x='date',y='revenue',figsize=(15,5))
```

# Plot trend and seasonal components 

```python
stl = STL(revenue_all['revenue'], period=30)
stl_model = stl.fit()
stl_model.plot()
plt.draw()
```

# Plot the autocorrelation and partial autocorrelation

```python
plot_acf(revenue_all['revenue'])
plt.draw()

plot_pacf(revenue_all['revenue'])
plt.draw()
```

# Augmented Dickey-Fuller unit root test

```python
adfuller_results = adfuller(revenue_all['revenue'])
if adfuller_results[1] < 0.05:
    print('Data set is stationary')
else:
    print('Data set is not stationary')
```
